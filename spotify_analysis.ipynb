```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tracks = pd.read_csv('C:/Users/CA Jagdish Dargar/Desktop/Python_spotify/tracks.csv')
tracks.head()
```

```python
#checking total null values in the dataset
pd.isnull(tracks).sum()
```

```python
# replaced all null values
tracks.fillna(value = 0,
          inplace = True)
```

```python
#checking total null values in the dataset
pd.isnull(tracks).sum()
```

```python
tracks.info()
```

```python
# 10 least popular songs
# sort_values function sorts a data frame in ascending or descending order of columns which is passed in the statement
sorted_least = tracks.sort_values('popularity', ascending = True).head(10)
sorted_least
```

```python
tracks.describe()
tracks.describe().transpose()
```

```python
# 10 most popular songs having more than 90 popularity
sorted_most = tracks.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
# using slicing operator
sorted_most[:10]     
```

```python
# replacing index column to release_date column
tracks.set_index('release_date', inplace = True)
tracks.index = pd.to_datetime(tracks.index)
tracks.head()
```

```python
# converting duration from miliseconds(ms) to seconds(s)
tracks['duration'] = tracks['duration_ms'].apply(lambda x: round(x/1000))
# dropping duration_ms column and added duration in seconds column
tracks.drop('duration_ms', inplace = True, axis = 1)
```

```python
tracks.tail(5)
```

```python
corr_df = tracks.drop(["key", "mode", "explicit"],axis=1).corr(method = "pearson")
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap = "inferno", linewidths = 1, linecolor = "Black")
heatmap.set_title("Correlation heatmap")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
# visualizing correlation between variables - loudness & energy. 
plt.figure(figsize=(10, 6))
sns.regplot(data = corr_df, y = 'loudness', x = 'energy', color = 'c').set(title = " Loudness vs Energy")
# we could find that it is a positive correlation b/w them. When loudness is more/ increases then energy increases
```

```python
# For getting a more clear visualization, we will take a part of this dataframe.
sample_df = tracks.sample(int(0.004 * len(tracks)))
len(sample_df)
```

```python
# When took a small part from the dataframe, the visualization is more clearer
plt.figure(figsize=(10, 6))
sns.regplot(data = sample_df, y = 'loudness', x = 'energy', color = 'c').set(title = " Loudness vs Energy Correlation")
## Loudness and Energy has positive correlation
```

```python
# another example of visualization for variables
plt.figure(figsize=(10, 6))
sns.regplot(data = sample_df, y = 'popularity', x = 'acousticness', color = 'g').set(title = " popularity vs acousticness Correlation")
# Popularity & acousticness have negative correlation
```

```python
genre = pd.read_csv("C:/Users/CA Jagdish Dargar/Desktop/Python_spotify/SpotifyFeatures.csv")
genre.head()
```
